```python
from tensorflow.keras.datasets import cifar10

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPool2D, Flatten, Dense, Dropout,Activation, BatchNormalization
from tensorflow.keras.losses import categorical_crossentropy
from tensorflow.keras.optimizers import RMSprop, Adam
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import pathlib
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from sklearn import datasets, svm, metrics
from sklearn.metrics import plot_confusion_matrix
from random import seed
from random import randint
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import confusion_matrix
import seaborn as sns; sns.set()
```

```python
trainModel = True #Make True if you wanna build your own model
# Model configuration
batch_size = 32
loss_function = categorical_crossentropy 
no_classes = 10
no_epochs = 50
#optimizer = RMSprop(lr=0.0001, decay=1e-6)
optimizer = Adam()
```

```python
# Load CIFAR-10 data
(input_train, target_train), (input_test, target_test) = cifar10.load_data()
```

```python
# Determine shape of the data
print(input_train.shape)
```

```python
input_train = np.array([cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) for image in input_train])
input_test = np.array([cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) for image in input_test])

input_train  = input_train/255
input_test  = input_test/255
```

```python
classes = { 
    0:'aeroplane',
    1:'automobile',
    2:'bird',
    3:'cat',
    4:'deer',
    5:'dog',
    6:'frog',
    7:'horse',
    8:'ship',
    9:'truck' 
}
print(target_train)
```

```python
fig, axes = plt.subplots(ncols=7, nrows=3, figsize=(17, 8))
index = 0
for i in range(3):
    for j in range(7):
        axes[i,j].set_title(classes[target_train[index][0]])
        axes[i,j].imshow(input_train[index], cmap='gray')
        axes[i,j].get_xaxis().set_visible(False)
        axes[i,j].get_yaxis().set_visible(False)
        index += 1
plt.show()
```

```python
one_hot_encoder = OneHotEncoder(sparse=False)
one_hot_encoder.fit(target_train)
target_train = one_hot_encoder.transform(target_train)
target_test = one_hot_encoder.transform(target_test)
print(target_train)
```

```python
input_train = input_train.reshape(input_train.shape[0], input_train.shape[1], input_train.shape[2], 1)
input_test = input_test.reshape(input_test.shape[0], input_test.shape[1], input_test.shape[2], 1)
input_shape = (input_train.shape[1], input_train.shape[2], 1)
```

```python
# Create the model
model = Sequential()
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=input_shape))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
model.add(MaxPool2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(no_classes, activation='softmax'))
model.summary()

modelRMS = Sequential()
modelRMS.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=input_shape))
modelRMS.add(MaxPool2D(pool_size=(2, 2)))
modelRMS.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))
modelRMS.add(MaxPool2D(pool_size=(2, 2)))
modelRMS.add(Conv2D(128, kernel_size=(3, 3), activation='relu'))
modelRMS.add(MaxPool2D(pool_size=(2, 2)))
modelRMS.add(Flatten())
modelRMS.add(Dense(256, activation='relu'))
modelRMS.add(Dense(128, activation='relu'))
modelRMS.add(Dense(no_classes, activation='softmax'))
modelRMS.summary()
```

```python
modelBig = Sequential()
modelBig.add(Conv2D(32, (3, 3), padding='same',
                    input_shape=input_shape))
modelBig.add(Activation('relu'))
modelBig.add(Conv2D(32, (3, 3)))
modelBig.add(Activation('relu'))
modelBig.add(MaxPool2D(pool_size=(2, 2)))
modelBig.add(Dropout(0.25))

modelBig.add(Conv2D(64, (3, 3), padding='same'))
modelBig.add(Activation('relu'))
modelBig.add(Conv2D(64, (3, 3)))
modelBig.add(Activation('relu'))
modelBig.add(MaxPool2D(pool_size=(2, 2)))
modelBig.add(Dropout(0.25))

modelBig.add(Flatten())
modelBig.add(Dense(512))
modelBig.add(Activation('relu'))
modelBig.add(Dropout(0.5))
modelBig.add(Dense(no_classes))
modelBig.add(Activation('softmax'))
modelBig.summary()

modelBigRMS = Sequential()
modelBigRMS.add(Conv2D(32, (3, 3), padding='same',
                    input_shape=input_shape))
modelBigRMS.add(Activation('relu'))
modelBigRMS.add(Conv2D(32, (3, 3)))
modelBigRMS.add(Activation('relu'))
modelBigRMS.add(MaxPool2D(pool_size=(2, 2)))
modelBigRMS.add(Dropout(0.25))

modelBigRMS.add(Conv2D(64, (3, 3), padding='same'))
modelBigRMS.add(Activation('relu'))
modelBigRMS.add(Conv2D(64, (3, 3)))
modelBigRMS.add(Activation('relu'))
modelBigRMS.add(MaxPool2D(pool_size=(2, 2)))
modelBigRMS.add(Dropout(0.25))

modelBigRMS.add(Flatten())
modelBigRMS.add(Dense(512))
modelBigRMS.add(Activation('relu'))
modelBigRMS.add(Dropout(0.5))
modelBigRMS.add(Dense(no_classes))
modelBigRMS.add(Activation('softmax'))
modelBigRMS.summary()
```

```python
# Compile the model
if(trainModel):
    model.compile(loss=loss_function,
              optimizer=Adam(),
              metrics=['accuracy'])
    modelBig.compile(loss=loss_function,
              optimizer=Adam(),
              metrics=['accuracy'])    
    modelRMS.compile(loss=loss_function,
              optimizer=Adam(),
              metrics=['accuracy'])
    modelBigRMS.compile(loss=loss_function,
              optimizer=Adam(),
              metrics=['accuracy'])    
    es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=3)

# Fit data to model
    print(input_train.shape)
    historyBig = modelBig.fit(input_train, target_train,
            batch_size=batch_size,
            epochs=no_epochs,
            validation_data=(input_test, target_test),
            callbacks=[es])
    history = model.fit(input_train, target_train,
            batch_size=batch_size,
            epochs=no_epochs,
            validation_data=(input_test, target_test),
            callbacks=[es])
    historyBigRMS = modelBigRMS.fit(input_train, target_train,
            batch_size=batch_size,
            epochs=no_epochs,
            validation_data=(input_test, target_test),
            callbacks=[es])
    historyRMS = modelRMS.fit(input_train, target_train,
            batch_size=batch_size,
            epochs=no_epochs,
            validation_data=(input_test, target_test),
            callbacks=[es])
```

```python
    # Generate generalization metrics
    score = model.evaluate(input_test, target_test, verbose=0)
    scoreBig = modelBig.evaluate(input_test, target_test, verbose=0)
    scoreRMS = modelRMS.evaluate(input_test, target_test, verbose=0)
    scoreRMSBig = modelBigRMS.evaluate(input_test, target_test, verbose=0)

    print(f'Test loss Adam: {score[0]} / Test accuracy Adam: {score[1]}')
    print(f'Test Big loss Adam: {scoreBig[0]} / Test Big accuracy Adam: {scoreBig[1]}')
    print(f'Test loss RMS: {scoreRMS[0]} / Test accuracy RMS: {scoreRMS[1]}')
    print(f'Test Big loss RMS: {scoreRMSBig[0]} / Test Big accuracy RMS: {scoreRMSBig[1]}')
    # Visualize history
    # Plot history: Loss
    plt.plot(history.history['val_loss'], '-', label = 'val_loss')
    plt.plot(history.history['loss'], '-',label = 'loss')
    plt.legend(loc=0)
    plt.title('Validation loss history Small Adam')
    plt.ylabel('Loss value')
    plt.xlabel('No. epoch')
    plt.savefig('loss Small Adam' + str(no_epochs) + '.png')
    plt.show()
    
    plt.plot(historyRMS.history['val_loss'], label = 'val_loss')
    plt.plot(historyRMS.history['loss'], label = 'loss')
    plt.legend(loc=0)
    plt.title('Validation loss history Small RMS')
    plt.ylabel('Loss value')
    plt.xlabel('No. epoch')
    plt.savefig('loss Small RMS' + str(no_epochs) + '.png')
    plt.show()
    
    plt.plot(historyBig.history['val_loss'], label = 'val_loss')
    plt.plot(historyBig.history['loss'], label = 'loss')
    plt.legend(loc=0)
    plt.title('Validation loss history Big Adam')
    plt.ylabel('Loss value')
    plt.xlabel('No. epoch')
    plt.savefig('loss Big Adam' + str(no_epochs) + '.png')
    plt.show()
    
    plt.plot(historyBigRMS.history['val_loss'], label = 'val_loss')
    plt.plot(historyBigRMS.history['loss'], label = 'loss')
    plt.legend(loc=0)
    plt.title('Validation loss history Big RMS')
    plt.ylabel('Loss value')
    plt.xlabel('No. epoch')
    plt.savefig('loss Big RMS' + str(no_epochs) + '.png')
    plt.show()
    
    plt.plot(historyBig.history['val_loss'], label = 'Big')
    plt.plot(history.history['val_loss'], label = 'Small')
    plt.legend(loc=0)
    plt.title('Validation loss Compare Adam')
    plt.ylabel('loss value (%)')
    plt.xlabel('No. epoch')
    plt.savefig('loss Compare Adam' + str(no_epochs) + '.png')
    plt.show()
    
    plt.plot(historyBigRMS.history['val_loss'], label = 'Big')
    plt.plot(historyRMS.history['val_loss'], label = 'Small')
    plt.legend(loc=0)
    plt.title('Validation loss Compare RMS')
    plt.ylabel('loss value (%)')
    plt.xlabel('No. epoch')
    plt.savefig('loss Compare RMS' + str(no_epochs) + '.png')
    plt.show()
    
    plt.plot(historyBig.history['val_loss'], label = 'Adam')
    plt.plot(historyBigRMS.history['val_loss'], label = 'RMS')
    plt.legend(loc=0)
    plt.title('Validation loss Compare Big')
    plt.ylabel('loss value (%)')
    plt.xlabel('No. epoch')
    plt.savefig('loss Compare Big' + str(no_epochs) + '.png')
    plt.show()
    
    plt.plot(history.history['val_loss'], label = 'Adam')
    plt.plot(historyRMS.history['val_loss'], label = 'RMS')
    plt.legend(loc=0)
    plt.title('Validation loss Compare Small')
    plt.ylabel('loss value (%)')
    plt.xlabel('No. epoch')
    plt.savefig('loss Compare Small' + str(no_epochs) + '.png')
    plt.show()
    
```

```python
    # Plot history: Accuracy
    plt.plot(history.history['val_accuracy'],label = 'val_accuracy')
    plt.plot(history.history['accuracy'],label = 'accuracy')
    plt.legend(loc=0)
    plt.title('Validation accuracy history Small Adam')
    plt.ylabel('accuracy value')
    plt.xlabel('No. epoch')
    plt.savefig('accuracy Small Adam' + str(no_epochs) + '.png')
    plt.show()
    
    plt.plot(historyRMS.history['val_accuracy'], label = 'val_accuracy')
    plt.plot(historyRMS.history['accuracy'], label = 'accuracy')
    plt.legend(loc=0)
    plt.title('Validation accuracy history Small RMS')
    plt.ylabel('accuracy value')
    plt.xlabel('No. epoch')
    plt.savefig('accuracy Small RMS' + str(no_epochs) + '.png')
    plt.show()
    
    plt.plot(historyBig.history['val_accuracy'], label = 'val_accuracy')
    plt.plot(historyBig.history['accuracy'], label = 'accuracy')
    plt.legend(loc=0)
    plt.title('Validation accuracy history Big Adam')
    plt.ylabel('accuracy value')
    plt.xlabel('No. epoch')
    plt.savefig('accuracy Big Adam' + str(no_epochs) + '.png')
    plt.show()
    
    plt.plot(historyBigRMS.history['val_accuracy'], label = 'val_accuracy')
    plt.plot(historyBigRMS.history['accuracy'], label = 'accuracy')
    plt.legend(loc=0)
    plt.title('Validation accuracy history Big RMS')
    plt.ylabel('accuracy value')
    plt.xlabel('No. epoch')
    plt.savefig('accuracy Big RMS' + str(no_epochs) + '.png')
    plt.show()
    
    plt.plot(historyBig.history['val_accuracy'], label = 'Big')
    plt.plot(history.history['val_accuracy'], label = 'Small')
    plt.legend(loc=0)
    plt.title('Validation accuracy Compare Adam')
    plt.ylabel('accuracy value (%)')
    plt.xlabel('No. epoch')
    plt.savefig('accuracy Compare Adam' + str(no_epochs) + '.png')
    plt.show()
    
    plt.plot(historyBigRMS.history['val_accuracy'], label = 'Big')
    plt.plot(historyRMS.history['val_accuracy'], label = 'Small')
    plt.legend(loc=0)
    plt.title('Validation accuracy Compare RMS')
    plt.ylabel('accuracy value (%)')
    plt.xlabel('No. epoch')
    plt.savefig('accuracy Compare RMS' + str(no_epochs) + '.png')
    plt.show()
    
    plt.plot(historyBig.history['val_accuracy'], label = 'Adam')
    plt.plot(historyBigRMS.history['val_accuracy'], label = 'RMS')
    plt.legend(loc=0)
    plt.title('Validation accuracy Compare Big')
    plt.ylabel('accuracy value (%)')
    plt.xlabel('No. epoch')
    plt.savefig('accuracy Compare Big' + str(no_epochs) + '.png')
    plt.show()
    
    plt.plot(history.history['val_accuracy'], label = 'Adam')
    plt.plot(historyRMS.history['val_accuracy'], label = 'Rms')
    plt.legend(loc=0)
    plt.title('Validation accuracy Compare Small')
    plt.ylabel('accuracy value (%)')
    plt.xlabel('No. epoch')
    plt.savefig('accuracy Compare Small' + str(no_epochs) + '.png')
    plt.show()
```

```python
labels = ['aeroplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck'] # labels
predictions = model.predict(input_test)
predictionsBig = modelBig.predict(input_test)

predictions = one_hot_encoder.inverse_transform(predictions)
predictionsBig = one_hot_encoder.inverse_transform(predictionsBig)

rounded_labels=np.argmax(target_test, axis=1)

cm = confusion_matrix(rounded_labels, predictions)


plt.figure(figsize=(9,9))
sns.heatmap(cm, cbar=False, xticklabels=labels, yticklabels=labels, fmt='d', annot=True, cmap=plt.cm.Blues)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.savefig('ConfusionSmallAdam' + str(no_epochs) + '.png')
plt.show()

cm = confusion_matrix(rounded_labels, predictionsBig)


plt.figure(figsize=(9,9))
sns.heatmap(cm, cbar=False, xticklabels=labels, yticklabels=labels, fmt='d', annot=True, cmap=plt.cm.Blues)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.savefig('ConfusionBigAdam' + str(no_epochs) + '.png')
plt.show()
```
